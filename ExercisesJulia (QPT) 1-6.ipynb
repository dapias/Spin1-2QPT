### Exercise 2:
###  Fix the number of up-spins and write down the Hamiltonian matrix for $\hat{H}_{ZZ}^{\rm{open}}$.  Use $L = 6$ and $N_{up} = L/2$. Show as output only the diagonal elements

Recall that $$\hat{H}_{ZZ}^{\rm{open}} = J \Delta \sum_{n=1}^{L-1} \hat{S}^z_n \hat{S}^z_{n+1} \, .$$ Let's take as an example the following configuration
$$|110100> \, .$$  The expected value of $\hat{H}_{ZZ}^{\rm{open}}$ is calculated as $$ <110100|\hat{H}_{ZZ}^{\rm{open}}|110100> = 
$$
$$  J \Delta  <110100| \,  \sum_{n=1}^{L-1} \hat{S}^z_n \hat{S}^z_{n+1} \, | 110100> = $$
$$ J \Delta  <110100| \, \hat{S}^z_1 \hat{S}^z_2 + \hat{S}^z_2 \hat{S}^z_3 + \hat{S}^z_3 \hat{S}^z_4 + \hat{S}^z_4 \hat{S}^z_5 + \hat{S}^z_5 \hat{S}^z_6 \, | 110100> = $$
$$ J \Delta  (1/4 - 1/4 - 1/4 - 1/4 + 1/4) = -\frac{J \Delta }{4} \, .$$

As it can be noted only the diagonal elements are different from zero.

$$H_{XXZ} = \sum_{n=1}^{L-1} \left[ J(S^{x}_n S^x_{n+1} + S^y_n S^y_{n+1}) + J \Delta S^z_n S^z_{n+1} \right]  $$

## (iii) Calculate the eigenvalues of $\hat{\Pi}$ and $\hat{R}^x_\pi$ and group them into a chart

For calculating the eigenvalues of $\hat{\Pi}$ I choose two elements of the basis such that they are related through the *"mirror"* operation. I pick the first one randmoly and then find its mirror. After that I compare its coefficients, if they have the same sign (and are reliable numbers in the sense that they are not near to the 0.0 of the machine) then the eigenvalue of $\hat{\Pi}$ is 1, if they differ by a minus sign then the eigenvalue of $\hat{\Pi}$ is $-1$.

```julia
par = parity_eigenvalues(v,states);
```

For calculating the eigenvalues of $\hat{R}^x_\pi$ we proceed in a similar way by picking two vectors of the basis related by the operator rotation and comparing its coefficients.

```julia
xrot = xrotation_eigenvalues(v,states);
```

Putting the three quantities associated to each state on a table

## Exercise 5: Make a plot of the value of $PR$ averaged over all the eigenstates of the XXZ model vs $\Delta$. Show two curves, one for the site-basis (blue) and the other one for the basis corresponding to the eigenstates of the XX part of the model (red). 

Let's call the basis given by the states $|1111100000>$, $|1111010000>$, $\ldots$: $|b_i>$. In Julia the eigenvectors of a matrix are arranged in the columns of a matrix. So if we arranged $|b_i>$ as a column vector, we have that the eigenvectors $|\Psi_j>$ of $H_{XXZ}$ can be written as:
$$ 
|\Psi> = D^T |b>
$$
At the same time, the eigenvectors $|\phi>$ of $H_{XX}$ are
$$
|\phi> = C^T |b>
$$
If we look for representing $|\Psi>$ in the $|\phi>$ basis we solve for $|b>$ in the last equation and substitute in the first one, obtaining
$$
|\Psi> = D^T (C^T)^{-1} |\phi>
$$
but $C$ is an orthogonal matrix (since $H_{XX}$ is a symmetric matrix).

```julia
base = basis(10,5)
delta = 0.0
ham_xx = h_xxz(base, d = delta)
statesxx = eigvecs(ham_xx);
```

```julia
delta, iprnew = averageipr_with_delta(1.0,20.0, basis(10,5),statesxx)
```

```julia
plot(delta, iprnew, "ro-")
plot(delta, ipr, "bo-")
plt[:ylabel](L"$<IPR>$")
plt[:xlabel](L"$\Delta$")
```
